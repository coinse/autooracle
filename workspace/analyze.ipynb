```python
from IPython.display import display
import pandas as pd 
import pickle
import os 

# Original
# 1. Get ground truth -> failing_tests_on_fixed는 <incorrect>, 나머지는 all <correct> - dataframe join
# 2. chat_reply parsing 해서 결과 얻기
# 3. dataframe - join
path = "./result/Lang-1b/generated_test/newTS"
# Read Evousite test metadata (columns=['dir', 'evo_relpath', 'evo_test_no', 'evo_test_src', 'evo_target_method'])

with open(os.path.join(path,"evo_test_df.pkl"),'rb') as f:
    evo_test_df = pickle.load(f)
# 
evo_test_df["ground_truth"] = "correct"

display(evo_test_df.head())
```

Parsing Chat_Reply

```python
from env import EvoD4jEnv
import re 

env = EvoD4jEnv("Lang", "1", "newTS")
chat_reply_dir = "./result/Lang-1b/generated_test/newTS/chat_reply"
chat_reply_list = os.listdir(chat_reply_dir)

for chat_reply_f in chat_reply_list:
    with open(os.path.join(chat_reply_dir, chat_reply_f),'r') as fr:
        data = fr.read()
    split_chat_reply_f = chat_reply_f.strip().split('_')
    dir = split_chat_reply_f[0].replace('.','/')
    test_no = split_chat_reply_f[1]
    if re.search("incorrect",data):
        evo_test_df.loc[ (evo_test_df["dir"] == dir) & (evo_test_df["evo_test_no"] == test_no), "chat_reply1" ] = "incorrect"
    elif re.search("correct",data):
        evo_test_df.loc[ (evo_test_df["dir"] == dir) & (evo_test_df["evo_test_no"] == test_no), "chat_reply1" ] = "correct" 

chat_reply2_dir = "./result/Lang-1b/generated_test/newTS/chat_reply2"
chat_reply2_list = os.listdir(chat_reply2_dir)

for chat_reply2_f in chat_reply2_list:
    with open(os.path.join(chat_reply2_dir, chat_reply2_f),'r') as fr:
        data = fr.read()
    split_chat_reply2_f = chat_reply2_f.strip().split('_')
    dir = split_chat_reply2_f[0].replace('.','/')
    test_no = split_chat_reply2_f[1]
    if re.search("incorrect",data):
        evo_test_df.loc[ (evo_test_df["dir"] == dir) & (evo_test_df["evo_test_no"] == test_no), "chat_reply2" ] = "incorrect"
    elif re.search("correct",data):
        evo_test_df.loc[ (evo_test_df["dir"] == dir) & (evo_test_df["evo_test_no"] == test_no), "chat_reply2" ] = "correct" 
evo_test_df.loc[:, ["evo_test_no","ground_truth","chat_reply1", "chat_reply2"]]
```
